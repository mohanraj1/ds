### Pascal Triangle

https://inst.eecs.berkeley.edu/~cs61a/fa11/61a-python/content/section_notes/week9/week09_sol.pdf



* Except for the first and last element in a line, rest of the entries are compute using the previous line
    * pascal(row, col) = pascal(row-1, col) + pascal(row-1, col-1)

* The following code only computes each cell

```python
def pascal(r, c, level):
    
    if r < 0 or c < 0:
        return 0
    
    if r == 0 and c == 0:
        return 1
    
    one_up = pascal(r-1, c, level+1)
    one_up_to_left = pascal(r-1, c-1, level+1)
    res = one_up + one_up_to_left
    print(("\t") * level, res)
    return res

pascal(4, 2, 0)
```

* Compute each line starting from lowest. Each subsequent line is built using previous line except for first and last element which is 1 - so simply add them
