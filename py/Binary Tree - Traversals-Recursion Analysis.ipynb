```python
def isIdentical(node1, node2, level, calledFrom):
    
    if node1 is None and node2 is None:
        return True
    
    if (node1 is None and node2 is not None) or (node1 is not None and node2 is None):
        return False
    
    if node1.value != node2.value:
        print(calledFrom + ('\t' * (level)) + str(node1.value) + ":" + str(node2.value) + "-not same")
        return False
    
    if node1 is not None and node2 is None:
        return False
    
    print(calledFrom + ('\t' * (level)) + str(node1.value) + ":" + str(node2.value))
    left = isIdentical(node1.left, node2.left, level+1, "L:")    
    right = isIdentical(node1.right, node2.right, level+1, "R:")
    
    result = left and right
    
    return result
```

1. In the first example: 6 and 5 are different
    * isIdentical(node1.left, node2.left, level+1, "L:") will return false
    * Since we are not returning like if (left == false) return False, the code stops going left because we returned but 
    * isIdentical(node1.right, node2.right, level+1, "R:") gets invoked
    * since nodes match, it keeps going and eventually stack unwinds 
    * and comes back to stack where we got False and at this point we return left and right
    
    * Optionally we can do the following
    * left = isIdentical(node1.left, node2.left, level+1, "L:")
    * if (left == false) return False. With this code won't proceed to isIdentical(node1.right, node2.right, level+1, "R:")
    

```python
node_values = [10, 6, 15, 4, 7, 12, 20, 3, None, None, 8, 11]
bst_root1 = build(node_values)
print(bst_root1)
node_values = [10, 5, 15, 4, 7, 12, 20, 3, None, None, 8, 11]
bst_root2 = build(node_values)
print(bst_root2)

print(isIdentical(bst_root1, bst_root2, 0, "O:"))

node_values = [10, 6, 15, 4, 7, 12, 20, 3, None, None, 8, 11]
bst_root1 = build(node_values)
print(bst_root1)
node_values = [10, 6, 15, 4, 7, 13, 20, 3, None, None, 8, 11]
bst_root2 = build(node_values)
print(bst_root2)

print(isIdentical(bst_root1, bst_root2, 0, "O:"))

bst_root1 = build(node_values)
print(bst_root1)
bst_root2 = build(node_values)
print(bst_root2)
print(isIdentical(bst_root1, bst_root2, 0, "O:"))
```

```python
def isIdenticalWithReturn(node1, node2, level, calledFrom):
    
    if node1 is None and node2 is None:
        return True
    
    if (node1 is None and node2 is not None) or (node1 is not None and node2 is None):
        return False
    
    if node1.value != node2.value:
        print(calledFrom + ('\t' * (level)) + str(node1.value) + ":" + str(node2.value) + "-not same")
        return False
    
    if node1 is not None and node2 is None:
        return False
    
    print(calledFrom + ('\t' * (level)) + str(node1.value) + ":" + str(node2.value))
    left = isIdenticalWithReturn(node1.left, node2.left, level+1, "L:")
    
    # usually we keep traversing towards left until we hit base condition or a condition fails 
    # if logic added so that we won't traverse right side
    if left == False:
        return False
    
    right = isIdenticalWithReturn(node1.right, node2.right, level+1, "R:")
    
    ''' #No need for this because there are no recursive calls
    if right == False:
        return False
    '''
    result = left and right
    
    return result
```

```python
node_values = [10, 6, 15, 4, 7, 12, 20, 3, None, None, 8, 11]
bst_root1 = build(node_values)
print(bst_root1)
node_values = [10, 5, 15, 4, 7, 12, 20, 3, None, None, 8, 11]
bst_root2 = build(node_values)
print(bst_root2)

print(isIdenticalWithReturn(bst_root1, bst_root2, 0, "O:"))

node_values = [10, 6, 15, 4, 7, 12, 20, 3, None, None, 8, 11, 14]
bst_root1 = build(node_values)
print(bst_root1)
node_values = [10, 6, 15, 4, 7, 13, 20, 3, None, None, 8, 11, 14]
bst_root2 = build(node_values)
print(bst_root2)

print(isIdenticalWithReturn(bst_root1, bst_root2, 0, "O:"))

bst_root1 = build(node_values)
print(bst_root1)
bst_root2 = build(node_values)
print(bst_root2)
print(isIdenticalWithReturn(bst_root1, bst_root2, 0, "O:"))
```
